The previous error "cannot import name 'gemma_runnable' from main" indicates that the Streamlit app cannot access variables directly from the Colab notebook's global scope. To fix this, we need to move the model initialization and `gemma_runnable` definition *inside* `app.py` itself. This will ensure that all necessary components are available when the Streamlit app starts.

```python
%%writefile app.py

import streamlit as st
import torch
from transformers import AutoProcessor, AutoModelForImageTextToText
from langchain_core.runnables import RunnableLambda

st.set_page_config(page_title="Colab Streamlit App with Gemma", page_icon="📊")

st.title("Hello from Colab Streamlit with Gemma! 👋")
st.write("This is a simple Streamlit app running inside Google Colab, integrated with your Gemma model.")

st.header("Your Gemma Model Demo")
st.write("Enter a text query below to get a response from the `gemma-4-E2B` model.")

# --- Model Initialization within Streamlit App ---
# These will only run once when the Streamlit app starts
@st.cache_resource
def load_model_and_processor():
    processor = AutoProcessor.from_pretrained("google/gemma-4-E2B")
    model = AutoModelForImageTextToText.from_pretrained("google/gemma-4-E2B")
    if torch.cuda.is_available():
        model.to('cuda')
    return processor, model

processor, model = load_model_and_processor()

# Define the function that will be wrapped by RunnableLambda
def generate_gemma_response(inputs: dict):
    image = inputs.get("image")
    text_prompt = inputs.get("text_prompt")

    if image and text_prompt:
        text_prompt_corrected = processor.tokenizer.image_token + " " + text_prompt
        model_inputs = processor(images=image, text=text_prompt_corrected, return_tensors="pt")
    elif text_prompt:
        model_inputs = processor(text=text_prompt, return_tensors="pt")
    else:
        raise ValueError("Either 'image' or 'text_prompt' must be provided in the input dictionary.")

    if torch.cuda.is_available():
        model_inputs = {k: v.to('cuda') for k, v in model_inputs.items()}
        # model.to('cuda') # Model is already moved to CUDA by @st.cache_resource
    else:
        model.to('cpu')

    generated_ids = model.generate(**model_inputs, max_new_tokens=100)
    generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return {"output": generated_text}

# Create a Langchain Runnable from the function
gemma_runnable = RunnableLambda(generate_gemma_response)

# --- Streamlit UI for interaction ---
user_input = st.text_input("Enter your query:", "What is the capital of France?")

if st.button("Get Response from Gemma"):
    if user_input:
        with st.spinner("Generating response..."):
            model_response = gemma_runnable.invoke({"text_prompt": user_input})
            st.write(f"**Gemma's Response:**")
            st.info(model_response.get("output", "No output generated."))
    else:
        st.warning("Please enter a query!")

st.sidebar.header("About")
st.sidebar.info("This app demonstrates Streamlit in Google Colab, integrated with the `gemma-4-E2B` model.")
st.sidebar.write("*(Note: For factual text-only responses, consider using instruction-tuned models like `google/gemma-2b-it` after gaining access.)*")
```

### Setting up a Dedicated Gemma Model Server with FastAPI

To create a truly persistent model server, we can use FastAPI. This will run the Gemma model in a separate process, and your Streamlit app will send requests to it. This decouples the model serving from the Streamlit UI, allowing the model to stay loaded and respond to queries even if the Streamlit app reloads.

First, we'll create a `model_server.py` file containing the FastAPI application. We'll also install the necessary libraries (`fastapi`, `uvicorn`, `python-multipart`).

```python
%%writefile model_server.py

import torch
from transformers import AutoProcessor, AutoModelForImageTextToText
from fastapi import FastAPI, UploadFile, File, Form
from PIL import Image
import io
import base64

app = FastAPI()

# --- Model Initialization ---
# This will run only once when the FastAPI application starts
processor = None
model = None

@app.on_event("startup")
async def load_model():
    global processor, model
    print("Loading Gemma model for FastAPI server...")
    processor = AutoProcessor.from_pretrained("google/gemma-4-E2B")
    model = AutoModelForImageTextToText.from_pretrained("google/gemma-4-E2B")
    if torch.cuda.is_available():
        model.to('cuda')
        print("Gemma model loaded onto GPU.")
    else:
        model.to('cpu')
        print("Gemma model loaded onto CPU.")

# --- Inference Endpoint ---
@app.post("/infer")
async def infer(text_prompt: str = Form(...), image_file: UploadFile = File(None)):
    input_image = None
    if image_file:
        image_bytes = await image_file.read()
        input_image = Image.open(io.BytesIO(image_bytes))

    if input_image and text_prompt:
        text_prompt_corrected = processor.tokenizer.image_token + " " + text_prompt
        model_inputs = processor(images=input_image, text=text_prompt_corrected, return_tensors="pt")
    elif text_prompt:
        model_inputs = processor(text=text_prompt, return_tensors="pt")
    else:
        return {"error": "Either 'image' or 'text_prompt' must be provided."}

    if torch.cuda.is_available():
        model_inputs = {k: v.to('cuda') for k, v in model_inputs.items()}
    else:
        model.to('cpu')

    generated_ids = model.generate(**model_inputs, max_new_tokens=100)
    generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return {"output": generated_text}
```

### Enhancing Streamlit App with OCR for Patient Data Extraction and Management

I'm updating the `app.py` file to include the following functionalities:

1.  **OCR Integration**: Using `easyocr` to extract text from uploaded medical images. This will attempt to find patient names, age, blood pressure, weight, and blood sugar values.
2.  **Pre-filling Input Fields**: If data is successfully extracted via OCR, the corresponding Streamlit input fields will be pre-filled, allowing for quick review and correction.
3.  **New 'Age' Field**: A new 'Age' field is added to the patient record both in the UI and the SQLite database schema.
4.  **Updated Database Schema**: The SQLite database (`patient_records.db`) will be updated to include the 'age' column. Existing records will default to `NULL` for age if the column was just added.
5.  **Robust Data Storage**: All vital signs, patient name, Gemma's analysis, and a timestamp will be stored in the SQLite database.
6.  **Historical Records Display**: The app will now display historical records for a given patient, including the new 'Age' field.

```python
%%writefile app.py

import streamlit as st
import requests
from PIL import Image
import io
import base64
import sqlite3
import pandas as pd
from datetime import datetime
import subprocess
import re

# Install necessary libraries silently
# Note: easyocr also requires opencv-python, which is a heavy dependency.
# Using opencv-python-headless for environments without a display.
subprocess.run([
    'pip', 'install',
    'easyocr',
    'opencv-python-headless'
], capture_output=True, text=True, check=True)

# Import after installation
import easyocr
import cv2
import numpy as np

# URL of our FastAPI model server
MODEL_SERVER_URL = "http://localhost:8000/infer"

# --- Database Functions ---
DB_NAME = "patient_records.db"

def init_db():
    conn = sqlite3.connect(DB_NAME)
    c = conn.cursor()
    c.execute("""
        CREATE TABLE IF NOT EXISTS patient_data (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            record_date TEXT NOT NULL,
            age INTEGER,
            blood_pressure TEXT,
            weight REAL,
            sugar REAL,
            image_path TEXT,
            gemma_analysis TEXT
        )
    """)
    conn.commit()
    conn.close()

def insert_patient_record(name, age, blood_pressure, weight, sugar, image_path, gemma_analysis):
    conn = sqlite3.connect(DB_NAME)
    c = conn.cursor()
    record_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    c.execute("""
        INSERT INTO patient_data (name, record_date, age, blood_pressure, weight, sugar, image_path, gemma_analysis)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    """, (name, record_date, age, blood_pressure, weight, sugar, image_path, gemma_analysis))
    conn.commit()
    conn.close()

def get_patient_records(name):
    conn = sqlite3.connect(DB_NAME)
    df = pd.read_sql_query(f"SELECT id, name, record_date, age, blood_pressure, weight, sugar, gemma_analysis FROM patient_data WHERE name = '{name}' ORDER BY record_date DESC", conn)
    conn.close()
    return df

# Initialize the database when the app starts
init_db()

# --- OCR Functions ---
@st.cache_resource
def get_ocr_reader():
    return easyocr.Reader(['en'])

reader = get_ocr_reader()

def extract_info_from_image(pil_image):
    # Convert PIL Image to OpenCV format
    open_cv_image = np.array(pil_image)
    # Convert RGB to BGR for OpenCV
    open_cv_image = open_cv_image[:, :, ::-1].copy()

    results = reader.readtext(open_cv_image)

    extracted_data = {
        "name": "",
        "age": None,
        "blood_pressure": "",
        "weight": None,
        "sugar": None,
    }

    text_lines = [res[1] for res in results]
    full_text = " ".join(text_lines).lower()

    # --- Extract Name (simple heuristic for demonstration, can be improved) ---
    name_match = re.search(r'(name|patient name|pt name):\s*([a-zA-Z\s\.]+)', full_text, re.IGNORECASE)
    if name_match:
        extracted_data["name"] = name_match.group(2).strip()
    else: # Try to find capitalized words followed by other text, assuming it's a name if no label
        potential_name_match = re.search(r'([A-Z][a-z]+\s+[A-Z][a-z]+)', full_text)
        if potential_name_match:
            extracted_data["name"] = potential_name_match.group(1).strip()

    # --- Extract Age ---
    age_match = re.search(r'(age|years old|y.o.|yr):\s*(\d+)', full_text, re.IGNORECASE)
    if age_match:
        extracted_data["age"] = int(age_match.group(2))

    # --- Extract Blood Pressure (e.g., 120/80, BP: 120/80) ---
    bp_match = re.search(r'(bp|blood pressure):\s*(\d{2,3}/\d{2,3})', full_text, re.IGNORECASE)
    if bp_match:
        extracted_data["blood_pressure"] = bp_match.group(2)
    else: # Try to find just the pattern if no explicit label
        bp_match = re.search(r'(\d{2,3}/\d{2,3})', full_text)
        if bp_match:
            extracted_data["blood_pressure"] = bp_match.group(1)


    # --- Extract Weight (e.g., Weight: 70 kg, 70kg, 70.5 lbs) ---
    weight_match = re.search(r'(weight|wt):\s*(\d+(\.\d+)?)\s*(kg|lbs|pounds)', full_text, re.IGNORECASE)
    if weight_match:
        extracted_data["weight"] = float(weight_match.group(2))
    else: # Try to find just number with 'kg' or 'lbs'
        weight_match = re.search(r'(\d+(\.\d+)?)\s*(kg|lbs)', full_text, re.IGNORECASE)
        if weight_match:
            extracted_data["weight"] = float(weight_match.group(1))


    # --- Extract Blood Sugar (e.g., Sugar: 100 mg/dL, Glucose: 100) ---
    sugar_match = re.search(r'(sugar|glucose):\s*(\d+(\.\d+)?)\s*(mg/dl|mg/dL)?', full_text, re.IGNORECASE)
    if sugar_match:
        extracted_data["sugar"] = float(sugar_match.group(2))
    else: # Try to find just a number followed by mg/dL
        sugar_match = re.search(r'(\d+(\.\d+)?)\s*mg/dl', full_text, re.IGNORECASE)
        if sugar_match:
            extracted_data["sugar"] = float(sugar_match.group(1))


    return extracted_data

st.set_page_config(page_title="Colab Streamlit App with External Gemma Server", page_icon="📊")

st.title("Medical Assistant with Gemma & Patient Records! 🩺")
st.write("Upload a medical image, input vital stats, and get Gemma's analysis, all saved to a patient database.")

st.header("Patient Information & Vitals")

# Initialize session state for patient info if not already present
if 'patient_name' not in st.session_state:
    st.session_state.patient_name = ""
if 'patient_age' not in st.session_state:
    st.session_state.patient_age = None
if 'blood_pressure' not in st.session_state:
    st.session_state.blood_pressure = ""
if 'patient_weight' not in st.session_state:
    st.session_state.patient_weight = None
if 'blood_sugar' not in st.session_state:
    st.session_state.blood_sugar = None


uploaded_file = st.file_uploader("Choose a medical image... (Optional for OCR pre-fill)", type=["jpg", "jpeg", "png"], key="image_uploader")

image = None
image_filename = ""
if uploaded_file is not None:
    try:
        image = Image.open(io.BytesIO(uploaded_file.getvalue()))
        image_filename = uploaded_file.name # Store filename for database reference
        st.image(image, caption='Uploaded Medical Image', use_column_width=True)

        st.subheader("Extracting info from image via OCR...")
        extracted_info = extract_info_from_image(image)

        # Update session state with extracted info
        if extracted_info["name"]:
            st.session_state.patient_name = extracted_info["name"]
        if extracted_info["age"] is not None:
            st.session_state.patient_age = extracted_info["age"]
        if extracted_info["blood_pressure"]:
            st.session_state.blood_pressure = extracted_info["blood_pressure"]
        if extracted_info["weight"] is not None:
            st.session_state.patient_weight = extracted_info["weight"]
        if extracted_info["sugar"] is not None:
            st.session_state.blood_sugar = extracted_info["sugar"]

        st.success("Information extracted from image (review and correct if necessary).")

    except Exception as e:
        st.error(f"Error loading or processing image for OCR: {e}")


patient_name = st.text_input("Patient Name:", value=st.session_state.patient_name, key="patient_name_input")
patient_age = st.number_input("Age:", min_value=0, max_value=150, value=st.session_state.patient_age, format="%d", key="age_input") # Added age
blood_pressure = st.text_input("Blood Pressure (e.g., 120/80):", value=st.session_state.blood_pressure, key="bp_input")
weight = st.number_input("Weight (kg):", min_value=0.0, value=st.session_state.patient_weight, format="%.2f", key="weight_input")
sugar = st.number_input("Blood Sugar (mg/dL):", min_value=0.0, value=st.session_state.blood_sugar, format="%.2f", key="sugar_input")


st.header("Medical Image Analysis with Gemma")

# --- Function to call the FastAPI server ---
def get_gemma_response_from_server(image: Image.Image = None, text_prompt: str = None):
    files = {}
    data = {"text_prompt": text_prompt}

    if image:
        img_byte_arr = io.BytesIO()
        # Try to save image in its original format, fallback to PNG
        try:
            image_format = image.format if image.format else 'PNG'
            image.save(img_byte_arr, format=image_format)
        except Exception:
            image.save(img_byte_arr, format='PNG')
        img_byte_arr = img_byte_arr.getvalue()
        files = {'image_file': ('image.png', img_byte_arr, 'image/png')}

    try:
        response = requests.post(MODEL_SERVER_URL, data=data, files=files)
        response.raise_for_status() # Raise an exception for HTTP errors
        return response.json().get("output", "Error: No output from server.")
    except requests.exceptions.ConnectionError:
        return "Error: Could not connect to the model server. Is it running?"
    except requests.exceptions.RequestException as e:
        return f"Error calling model server: {e}"


user_input = st.text_area("Enter your medical question or prompt for Gemma:", "Describe what you see in this image, focusing on any abnormalities.", key="gemma_prompt_input")

if st.button("Get Analysis & Save Record"):
    if not patient_name:
        st.warning("Please enter a patient name.")
    elif image is None and not user_input:
        st.warning("Please upload an image or enter a question/prompt for Gemma!")
    else:
        with st.spinner("Analyzing with Gemma Server and saving record..."):
            gemma_analysis = get_gemma_response_from_server(image=image, text_prompt=user_input)

            # For simplicity, we'll store the image path as its filename if uploaded, otherwise empty
            # In a real app, you'd save the image bytes to DB or a file storage and store its reference.
            image_db_path = image_filename if image else ""

            insert_patient_record(patient_name, patient_age, blood_pressure, weight, sugar, image_db_path, gemma_analysis)
            st.success("Patient record saved successfully!")
            st.write(f"**Gemma's Analysis (from server):**")
            st.info(gemma_analysis)

st.subheader(f"Historical Records for {patient_name}")

if patient_name:
    historical_df = get_patient_records(patient_name)
    if not historical_df.empty:
        st.dataframe(historical_df)
    else:
        st.info(f"No historical records found for {patient_name}.")
else:
    st.info("Enter a patient name to view historical records.")

st.sidebar.header("About")
st.sidebar.info("This app demonstrates Streamlit in Google Colab, leveraging a separate FastAPI server for the `google/gemma-4-E2B` model and an SQLite database for patient records. Now includes OCR for pre-filling vitals!")
st.sidebar.write("*(Note: `gemma-4-E2B` is a general multi-modal model and may not provide medically accurate diagnoses. For specific medical tasks, specialized models are required. Also, for factual text-only responses, consider using instruction-tuned models like `google/gemma-2b-it` after gaining access.)*")
```
